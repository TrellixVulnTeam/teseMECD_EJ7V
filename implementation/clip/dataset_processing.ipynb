```python
IMAGE_EXTENSION = ".tif"

vqaLR_dataset = datasets.load_dataset("csv", data_files={"train": "/content/drive/My Drive/Thesis Datasets/RSVQA_LR/Original/traindf.csv", "test": "/content/drive/My Drive/Thesis Datasets/RSVQA_LR/Original/testdf.csv", "validation": "/content/drive/My Drive/Thesis Datasets/RSVQA_LR/Original/valdf.csv"})
vqaLR_dataset = datasets.DatasetDict({"train": vqaLR_dataset["train"], # remove this when processing entire dataset
                                      "test": vqaLR_dataset["test"],
                                      "validation":vqaLR_dataset["validation"]})
vqaLR_dataset
```

```python
vqaLR_dataset = vqaLR_dataset.cast_column("img_id", datasets.Value("string"))
vqaLR_dataset = vqaLR_dataset.rename_column("answer", "labels")
vqaLR_dataset
```

```python
features = vqaLR_dataset["train"].features
features
```

```python
vqaLR_dataset_df = vqaLR_dataset["train"].to_pandas()
vqaLR_dataset_df["labels"].value_counts(normalize=True).sort_index()
```

```python
class_weights = (1 - (vqaLR_dataset_df["labels"].value_counts(normalize=True).sort_index())/ len(vqaLR_dataset_df)).values
class_weights = torch.from_numpy(class_weights).float()
print(class_weights)
```

```python
#tokenize the data
#ver encoders single stream e dual stream (o clip é dual stream)
processed = {}
processor = transformers.CLIPProcessor.from_pretrained("flax-community/clip-rsicd-v2")

processed["train"] = processor(vqaLR_dataset["train"]["question"], padding=True, return_tensors="pt")
processed["test"] = processor(vqaLR_dataset["test"]["question"], padding=True, return_tensors="pt")
processed["validation"] = processor(vqaLR_dataset["validation"]["question"], padding=True, return_tensors="pt")
processed
```

```python
vqaLR_dataset["train"] = vqaLR_dataset["train"].map(lambda example, idx: {"input_ids": processed["train"]["input_ids"][idx], "attention_mask": processed["train"]["attention_mask"][idx]}, with_indices=True)
vqaLR_dataset["test"] = vqaLR_dataset["test"].map(lambda example, idx: {"input_ids": processed["test"]["input_ids"][idx], "attention_mask": processed["test"]["attention_mask"][idx]}, with_indices=True)
vqaLR_dataset["validation"] = vqaLR_dataset["validation"].map(lambda example, idx: {"input_ids": processed["validation"]["input_ids"][idx], "attention_mask": processed["validation"]["attention_mask"][idx]}, with_indices=True)
vqaLR_dataset
```

```python
stop = False
for split in processed:
  for i in range(processed[split]["input_ids"].size()[0]):
    if torch.equal(processed[split]["input_ids"][i], torch.tensor(vqaLR_dataset[split][i]["input_ids"])) and\
      torch.equal(processed[split]["attention_mask"][i], torch.tensor(vqaLR_dataset[split][i]["attention_mask"])):
      continue
    else:
      print("not matching mapping at index", i)
      stop = True
      break
  if stop:
    break
```
